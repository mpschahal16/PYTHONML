```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as pit
```

```python
x,y=make_blobs(n_samples=400,centers=5,cluster_std=0.6,random_state=0)
pit.scatter(x[:,0],x[:,1],s=50)
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=5)
km.fit(x)
pre=km.predict(x)
pit.scatter(x[:,0],x[:,1],c=pre,cmap='winter')

centers=km.cluster_centers_
pit.scatter(centers[:,0],centers[:,1],c='red',s=200,alpha=0.6)
```

```python
cluster_range=range(1,15)
cluster_errors=[]

for k in cluster_range:
    km=KMeans(k)
    km.fit(x)
    cluster_errors.append(km.inertia_)
```

```python
import pandas as pd
df=pd.DataFrame({'num_clusters':cluster_range,'cluster_error':cluster_errors})
df
```

```python
pit.plot(cluster_range,cluster_errors,marker='*')

pit.xlabel('no. of cluster')
pit.ylabel('wcss error')
pit.show()


```
